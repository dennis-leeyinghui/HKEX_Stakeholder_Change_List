# Loading Websites + Parsing Data

```python
# Search Between 23 Feb, 2016 - 2017
# url that we are scraping

'''
*********************************
Setting website
*********************************

http = r"http://sdinotice.hkex.com.hk/di/NSAllFormDateList.aspx?sa1=da&scsd=23%2f02%2f2016&sced=23%2f02%2f2017&src=MAIN&lang=ZH&pg="
page = 1

url = http + str(page)
'''
http = r"http://sdinotice.hkex.com.hk/di/NSAllFormDateList.aspx?sa1=da&scsd=23%2f02%2f2016&sced=23%2f02%2f2017&src=MAIN&lang=ZH&pg="


for page in range(1,5):

    html = urllib.request.urlopen((http+str(page)))
    soup = BeautifulSoup(html)

    transactionList = []

    '''
    *********************************
    Locating "td" tag
    *********************************
    '''

    for name in (soup.findAll("td", "tbCell")):     

        name = name.contents[0]
        name = name.string.extract()  

        global new_name
        new_name = name
        transactionList.append([new_name])

    list1 = []
    list2 = []

    list1 = [transactionList[x+2:x+10] for x in range(-2, len(transactionList),10)]

    list2 = list1[:]

    '''
    *********************************
    CSV Writer
    *********************************
    '''
    with open('names.csv', 'a', encoding='utf-8-sig') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(['Date', 'Corporation', 'Shareholder', 'Code', 'Shares Involved', 'Average Price', 'Total Shares', '% of issued shared'])

        try:
            for x in range(0, len(list2)-1):
                data = [(list2[x][0][0],list2[x][1][0],list2[x][2][0],list2[x][3][0],list2[x][4][0],list2[x][5][0],list2[x][6][0],list2[x][7][0])]    
                print (data)
                writer.writerows(data)

        except IOError as err:
            pass

csvfile.close()
print (csvfile)


```
